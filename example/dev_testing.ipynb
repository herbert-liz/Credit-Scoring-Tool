## Subpaquete Data
Se realizan pruebas tanto para valores correctos como para valores incorrectos en la función `validate_dataframe` (y las funciones de las que depende) del módulo `loader`. Esta función se encarga de validar que el objeto es un DataFrame de pandas, que no esté vacío y que no tenga columnas duplicadas.

### Modulo loader

#### load_csv
Función para cargar datos de un archivo csv

```python
from creditScoring.data.loader import load_csv
datos = load_csv('application_train.csv')
```

#### ensure_dataframe
Función para validar que el objeto es un DataFrame de pandas

```python
# Revisión de error
from creditScoring.data.loader import ensure_dataframe
ensure_dataframe(1)
```

```python
# Revisión de valor correcto
ensure_dataframe(datos)
```

#### check_empty
Función para validar que el DataFrame no esté vacío

```python
from creditScoring.data.loader import check_empty
datosVacio = pd.DataFrame()
check_empty(datosVacio)
```

```python
# Revisión de valor correcto
check_empty(datos)
```

#### check_duplicate_columns
Función para validar que el DataFrame no tenga columnas duplicadas

```python
from creditScoring.data.loader import check_duplicate_columns

ejemploDf = pd.DataFrame({'col1': [1, 2, 3],'col2': [4, 5, 6],'col1_2': [7, 8, 9]})
ejemploDf = ejemploDf.rename(columns={'col1_2': 'col1'})
check_duplicate_columns(ejemploDf)
```

```python
# Revisión valor correcto
check_duplicate_columns(datos)
```

#### validate_dataframe
Función para validar que el objeto es un DataFrame de pandas, que no esté vacío y que no tenga columnas duplicadas

```python
# Revisión de valor incorrecto
cst.validate_dataframe(datosVacio)
```

```python
# Revisión de valor correcto
cst.validate_dataframe(datos)
```
